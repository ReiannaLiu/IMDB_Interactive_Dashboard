```python
from config import api_key
import requests
import csv
import pandas as pd
import pprint
```

```python
# Get data from https://rapidapi.com/
# title/get-sound-tracks


# load csv file
titles = pd.read_csv("./Resources/IMDB_API_adventure100.csv")



url = "https://imdb8.p.rapidapi.com/title/get-sound-tracks"

title_id = titles['title'][0]
# create empty list to store data
soundtracks = []

querystring = {"tconst":title_id}
headers = {
	"X-RapidAPI-Key": api_key,
	"X-RapidAPI-Host": "imdb8.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)


pprint.pprint(response.json())
```

```python

# export the data in pf
db = pd.DataFrame(response.json())
# set haeader as 'title'


# remove the prefix

# export to csv
db.to_csv('./dataset/IMDB_API_adventure100_soundtrack.csv', index=False)
```
