```python
from config import api_key
import requests
import csv
import pandas as pd
import pprint
import time
```

```python
# Get data from https://rapidapi.com/
# title/get-sound-tracks


# load csv file
titles = pd.read_csv("./Resources/IMDB_API_adventure100.csv")



url = "https://imdb8.p.rapidapi.com/title/get-sound-tracks"


# create empty list to store data
soundtracks = []
moivesinfo = []
for title in titles['title']:
	time.sleep(5)
	querystring = {"tconst":title}
	headers = {
		'x-rapidapi-key': api_key,
		'x-rapidapi-host': "imdb8.p.rapidapi.com"
		}

	response = requests.get(url, headers=headers, params=querystring)
	# pprint.pprint(response.json())
	
# loop through the response and append the data to the list as dictionary	
	try:
		
		for soundtrack in response.json()['soundtracks']:
		# pprint.pprint(soundtrack)
			soundtrackinfo = {}
			soundtrackinfo['id']=title
			soundtrackinfo['soundtracks_name'] = soundtrack['name']
			soundtrackinfo['soundtracks_id']=soundtrack['id']
			soundtrackinfo['soundtracks_comment'] = soundtrack['comment']
			soundtracks.append(soundtrack)
# wait for 5 seconds to avoid the API limit
		
	except Exception as e:
		print(e)
		print(f"Can't find the soundtrack for {title}")
		continue
	try:
			moive = response.json()['base']
			# pprint.pprint(moive)
			moiveinfo = {}
			moiveinfo['id']=title
			moiveinfo['image_url'] = moive['image']['url']
			moiveinfo['runningTimeInMinutes']=moive['runningTimeInMinutes']
			moiveinfo['title'] = moive['title']
			moiveinfo['titleType']=moive['titleType']
			moiveinfo['year']=moive['year']
			soundtracks.append(moiveinfo)
	except Exception as e:
		print(e)
		print(f"Can't find the moiveinfo for {title}")
		continue
	print("------------------------------------")
	pprint.pprint(soundtracks)
	pprint.pprint(moivesinfo)
	print("------------------------------------")
# create a dataframe from the list

soundtracks_df = pd.DataFrame(soundtracks)
moivesinfo_df = pd.DataFrame(moivesinfo)

# save the dataframe to csv file
soundtracks_df.to_csv("./Resources/IMDB_API_adventure100_soundtrack.csv", index=False)
moivesinfo_df.to_csv("./Resources/IMDB_API_adventure100_moiveinfo.csv", index=False)


```
