```python
from config import api_key
import requests
import csv
import pandas as pd
import pprint
import time
```

```python
# Get data from https://rapidapi.com/
# title/get-sound-tracks


# load csv file
titles = pd.read_csv("./Resources/IMDB_API_adventure100.csv")



url = "https://imdb8.p.rapidapi.com/title/get-sound-tracks"


# create empty list to store data
soundtracks = []
moivesinfo = []

# soundtracksraw = []
# moivesinforaw = []
```

```python
for title in titles['title']:
	time.sleep(1)
	querystring = {"tconst":title}
	headers = {
		'x-rapidapi-key': api_key,
		'x-rapidapi-host': "imdb8.p.rapidapi.com"
		}

	response = requests.get(url, headers=headers, params=querystring)
	# pprint.pprint(response.json())
	
# loop through the response and append the data to the list as dictionary	
	
	
	moive = response.json()['base']
	
	moiveinfo = {}
	moiveinfo['id']=title
	try:
		moiveinfo['image_url'] = moive['image']['url']
	except Exception as e:
		print(e)
		print(f"Can't find the url for {title}")
		continue
	try:
		moiveinfo['runningTimeInMinutes']=moive['runningTimeInMinutes']
		moiveinfo['title'] = moive['title']
	except Exception as e:
		print(e)
		print(f"Can't find the title or length for {title}")
		continue
	try:
		moiveinfo['titleType']=moive['titleType']
		moiveinfo['year']=moive['year']
		pprint.pprint(moivesinfo)
		moivesinfo.append(moiveinfo)
	except Exception as e:
		print(e)
		print(f"Can't find the titleTypr or year for {title}")
		continue

	if 'soundtracks' not in response.json():
		print(f"No soundtrack for {title}")
	else:
		for soundtrack in response.json()['soundtracks']:
		# pprint.pprint(soundtrack)
			soundtrackinfo = {}
			soundtrackinfo['id']=title
			try:
				soundtrackinfo['soundtracks_id']=soundtrack['id']
				soundtrackinfo['soundtracks_name'] = soundtrack['name']
			except Exception as e:
				print(e)
				print(f"Can't find the soundtrack base info for {title}")
				continue			
			try:
				soundtrackinfo['soundtracks_comment'] = soundtrack['comment']
			except Exception as e:
				print(e)
				print(f"Can't find the soundtrack comment  for {title}")
				continue
		
			
			soundtracks.append(soundtrackinfo)
			pprint.pprint(soundtrackinfo)

# wait for 5 seconds to avoid the API limit

	
	
	
	

```

```python
# create a dataframe from the list

soundtracks_df = pd.DataFrame(soundtracks)
# reorder the columns
soundtracks_df = soundtracks_df[['id', 'soundtracks_id', 'soundtracks_name', 'soundtracks_comment']]



moivesinfo_df = pd.DataFrame(moivesinfo)
# #unqiue values
moivesinfo_df = moivesinfo_df.drop_duplicates(subset=['id'])
# reorder the columns
moivesinfo_df = moivesinfo_df[['id', 'title', 'titleType', 'year', 'runningTimeInMinutes','image_url']]
# save the dataframe to csv file

soundtracks_df.to_csv("./Resources/IMDB_API_adventure100_soundtrack.csv", index=False)
moivesinfo_df.to_csv("./Resources/IMDB_API_adventure100_moiveinfo.csv", index=False)
soundtracks_df.to_json("./Resources/IMDB_API_adventure100_soundtrack.json", orient="records")
moivesinfo_df.to_json("./Resources/IMDB_API_adventure100_moiveinfo.json", orient="records")
```
