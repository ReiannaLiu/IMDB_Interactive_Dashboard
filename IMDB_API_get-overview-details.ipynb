```python
from config import api_key
import requests
import csv
import pandas as pd
import pprint
import time
```

```python
# Get data from https://rapidapi.com/
# title/get-sound-tracks


# load csv file
titles = pd.read_csv("./Resources/IMDB_API_adventure100.csv")



url = "https://imdb8.p.rapidapi.com/title/get-overview-details"


# create empty list to store data
# soundtracks = []
moivesinfos = []

# soundtracksraw = []
# moivesinforaw = []
```

```python
for title in titles['title']:
	time.sleep(1)
	querystring = {"tconst":title}
	headers = {
		'x-rapidapi-key': api_key,
		'x-rapidapi-host': "imdb8.p.rapidapi.com"
		}

	response = requests.get(url, headers=headers, params=querystring)
	# pprint.pprint(response.json())
	
# loop through the response and append the data to the list as dictionary	
	
	
	moive = response.json()
	
	moiveinfo = {}
	moiveinfo['id']=title
	if 'title' in response.json():
		if 'title' in response.json()['title']:
			moiveinfo['title'] = moive['title']['title']
		# else:
		# 	moiveinfo['title']="NA"


		if 'titleType' in response.json()['title']:
			moiveinfo['titleType']=moive['title']['titleType']
		# else:
		# 	moiveinfo['titleType']='NA'


		if 'runningTimeInMinutes' in response.json()['title']:
			moiveinfo['runningTimeInMinutes']=moive['title']['runningTimeInMinutes']
		# else:
		# 	moiveinfo['runningTimeInMinutes']=0
		

		if 'year' in response.json()['title']:
			moiveinfo['year']=moive['title']['year']
		# else:
		# 	moiveinfo['year']=0


		if 'image' in response.json()['title']:
			if 'url' in response.json()['title']['image']:
				moiveinfo['image_url'] = moive['title']['image']['url']
		# 	else:
		# 		moiveinfo['image_url']='NA'
		# else:
		# 	moiveinfo['image_url']='NA'

		
	# else:
	# 	moiveinfo['title']= 'NA'
	# 	moiveinfo['titleType']='NA'
	# 	moiveinfo['runningTimeInMinutes']='NA'
	# 	moiveinfo['image_url']='NA'
	# 	moiveinfo['year']='NA'

	if 'genres' in response.json():
		moiveinfo['genres'] = moive['genres']
	# else:
	# 	moiveinfo['genres']=[]


	if 'releaseDate' in response.json():
		moiveinfo['releaseDate']=moive['releaseDate']
	# else:
	# 	moiveinfo['releaseDate']='NA'
	
	moivesinfos.append(moiveinfo)
	pprint.pprint(moiveinfo)
```

```python
# create a dataframe from the list

moivesinfo_df = pd.DataFrame(moivesinfos)


moivesinfo_df.head()
```

```python
# reorder the columns
moivesinfo_df = moivesinfo_df[['id', 'title', 'titleType','genres', 'year', 'releaseDate','runningTimeInMinutes','image_url']]
# save the dataframe to csv file


moivesinfo_df.to_csv("./Resources/IMDB_API_adventure100_moiveinfo.csv", index=False)

moivesinfo_df.to_json("./Resources/IMDB_API_adventure100_moiveinfo.json", orient="records")
```
